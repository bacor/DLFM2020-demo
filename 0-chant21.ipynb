## 2. Volpiano

````python
import chant21
from music21 import converter # Note: not tested with 6.1 yet!
````

Chant21's Volpiano format is called `cantus`. You can specify the lyrics after a slash:

````python
ch = converter.parse('1---f-g--h---g--f--h---3/Abra cadabra', format='cantus')
ch.show('html')
````

````python
ch.show()
````

````python
ch.show('text')
````

## 3. gabc

````python
ch2 = converter.parse('gabc: (c2) a(f)b(g) (::)')
ch2.show('html')
````

````python
ch2.show('text')
````

## 4. Display options

````python
from chant21.examples import kyrie as kyrieFilename
kyrie = converter.parse(kyrieFilename, format='gabc')
kyrie.show('html', showOptions=True)
````
